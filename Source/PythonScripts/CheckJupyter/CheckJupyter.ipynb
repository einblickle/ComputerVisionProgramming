## Import

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import plotly.express as px
```

## Load Image

```python
img = cv2.imread(r'Imgs\TestImg.jpg', cv2.IMREAD_GRAYSCALE)
```

## Plotly Display Image

```python
fig = px.imshow(img, color_continuous_scale='gray')
fig.update_coloraxes(showscale=False)
fig.show()
```

## Matplotlib display

```python
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

## Open CV Transformation

```python
center = (img.shape[1]//2, img.shape[0]//2)
angle = -50
scale = 0.6
rot_mat = cv2.getRotationMatrix2D( center, angle, scale )
img_warp = cv2.warpAffine(img, rot_mat, (img.shape[1], img.shape[0]))

plt.imshow(img_warp, cmap='gray')
plt.axis('off')
plt.show()
```
