# Übung Digitale Bilder

```python
# package import
import numpy as np
import matplotlib.pyplot as plt
import cv2
import plotly.express as px
```

### Einlesen und Ausgabe der Datentypen und der Bildgröße
#### Graustufenbild

```python
imageLighthouse = cv2.imread(r'images/lighthouse.bmp', cv2.IMREAD_GRAYSCALE)
# Zur kontrolle Datentyp des Bildes Ausgeben
#Datentyp:
imageDataType = imageLighthouse.dtype
#Bildgröße
imageShape = imageLighthouse.shape

# Ausgabe der Daten
print(f'Datentyp des Bildes: {imageDataType}')
print(f'Größe des Bildes: {imageShape}')

# Anzeige des Bildes
fig = px.imshow(imageLighthouse, color_continuous_scale='gray')
fig.update_layout(
    margin=dict(l=0, r=0, t=0, b=0),
    width=imageLighthouse.shape[1],
    height=imageLighthouse.shape[0]
)
fig.show()
```

### Hellster und Dunkelster Pixel
Wenden wir uns nun wieder dem Graustufenbild des Leuchtturms zu `imageLighthouse`. Finden Sie im Bild den hellsten bzw. den dunkelsten Pixel. Geben Sie sowohl den Grauwert als auch die Position dieses Pixels an.
Das Bild enthält mehrere hellste/dunkelste Pixel mit gleicher Helligkeit. Es genügt wenn Sie die Position eines Pixels angeben. Beschreiben sie wo sich dieser Pixel befindet.

```python
# Enter your Code here:

# größter Helligkeitswert:
intensityBright = np.max(imageLighthouse)
# kleinster Helligkeitswert:
intensityDark = np.min(imageLighthouse)

# Position des hellsten Pixels
posOfBrightPixelList = np.where(imageLighthouse==intensityBright)
posOfBrightPixel = np.array([posOfBrightPixelList[0][0], posOfBrightPixelList[1][0]])

# Position des dunkelsten Pixels
posOfDarkPixelList = np.where(imageLighthouse==intensityDark)
posOfDarkPixel = np.array([posOfDarkPixelList[0][0], posOfDarkPixelList[1][0]])

print('Hellster Pixel:')
print(f'Grauwert: {intensityBright}')
print(f'Position (y,x): {posOfBrightPixel}')

print('Dunkelster Pixel:')
print(f'Grauwert: {intensityDark}')
print(f'Position (y,x): {posOfDarkPixel}')
#
```

### Bildausschnitt Wählen / Freistellen
Im letzten Teil sollen rechteckige Bildauschnitte so festgelegt werden, dass

* im Bild `imageLighthouse` nur das Fernglas ausgeschnitten wird
und
* im Bild `imageButterfly` nur der Schmetterling
Diese Bildausschnitte nennt man ROI (Region Of Interest)

Bestimmen Sie hierzu die Koordinate der Eckpunkte des Bildausschnittes und schränken sie den Bildbereich auf diese Koordinaten ein. Speichern sie die Daten in `imageLighthouse_ROI` und `imageButterfly_ROI` ab

```python
# Anzeige der Bilder

imageButterfly_ROI = imageButterfly[81:440, 150:670]
imageLighthouse_ROI = imageLighthouse[115:300, 400:470]

# Anzeige des Bildes
fig = px.imshow(imageLighthouse_ROI, color_continuous_scale='gray')
fig.update_layout(
    margin=dict(l=0, r=0, t=0, b=0),
    width=imageLighthouse_ROI.shape[1],
    height=imageLighthouse_ROI.shape[0]
)
fig.show()

# Anzeige des Bildes
fig = px.imshow(imageButterfly_ROI)
fig.update_layout(
    margin=dict(l=0, r=0, t=0, b=0),
    width=imageButterfly_ROI.shape[1],
    height=imageButterfly_ROI.shape[0]
)
fig.show()
```
